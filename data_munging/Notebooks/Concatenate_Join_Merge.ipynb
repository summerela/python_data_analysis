## Concatenate, Join and Merge Data Frames
We can combine dataframes in similar fashion to SQL joins, which are based on set theory.

### Concatenate Data Frames
Concatenating dataframes joins them together either by adding one dataframe on to the end of the other, or side by side. 

This is the perfect way to quickly add together two dataframes of the same length. 

Let's create some dummy dataframes: 

```python
import pandas as pd

raw_data = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
df1 = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df1
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df2 = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df2
```

`pd.concat([list of DataFrames], axis=0, join='outer', ignore_index=False)`

The concat() method takes a list of dataframes and the following arguments: 
- axis = 0 to join along rows, axis=1 join along columns
- join = 'outer', 'left', 'right'
- ignore_index = True to keep original row labels



### Concat columns

```python
pd.concat([df1,df2], axis=1, join="outer")
```

### Concat rows

```python
pd.concat([df1,df2], axis=0, join="inner")
```

## Merge "join" two Data Frames
![joins](https://www.dofactory.com/Images/sql-joins.png)

By default, the merge() function looks from common column names between the two tables to merge on. 

We can use the `on` argument to specify particular columns to merge on. If there are no common column names, we can use the `left_on` and `right_on` arguments to tell pandas how we want to combine information: 

#### Left Join
In a left join, all information from the left table is retained. If the right table does not have a match, additional columns will be filled in with `NaN`. 

```python
result = pd.merge(df1, df2, on='subject_id', how="left")
result.head()
```

#### Inner Join (default)
In an innder join, only rows that appear in both tables are retained. 

```python
# inner join is default
result1 = pd.merge(df1, df2, on='subject_id', how="inner") 
result1.head()
```

#### Outer Join
In an outer join, all rows from both tables are returned. Where a match was not found between them, columns will be filled in with `NaN`. 

```python
result2 = pd.merge(df1, df2, on='subject_id', how='outer') 
result2.head()
```

#### Right Join
This is pretty much the opposite of the left join, where all rows from the right table will be retained: 

```python
result3 = pd.merge(df1, df2, on='subject_id', how='right') 
result3.head()
```

# Data Exploration

## Read in dataframe

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
import pandas as pd
df = pd.read_csv(url)
print(df.head())
```

## Dataframe Summary Statistics

```python
df.describe()
```

```python
import pandas_profiling
pandas_profiling.ProfileReport(df)
```

```python
# view variables rejected due to high correlation
profile = pandas_profiling.ProfileReport(df)
rejected_variables = profile.get_rejected_variables(threshold=0.9)
```

```python
# find missing values
length = len(df[' Never-married'])
count = df[' Never-married'].count()
number_of_missing_values =  length - count
pct_of_missing_values = float(number_of_missing_values / length)
pct_of_missing_values = "{0:.1f}%".format(pct_of_missing_values*100)
print(pct_of_missing_values)
```
